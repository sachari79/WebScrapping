```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from wait_for import wait_for
```

```python
driver = webdriver.Firefox()
```

```python
driver.implicitly_wait(10)
```

```python
driver.get("http://www.dallascad.org/SearchAddr.aspx")
```

```python
class wait_for_page_load(object):

    def __init__(self, browser):
        self.browser = browser

    def __enter__(self):
        self.old_page = self.browser.find_element_by_tag_name('html')

    def page_has_loaded(self):
        new_page = self.browser.find_element_by_tag_name('html')
        return new_page.id != self.old_page.id

    def __exit__(self, *_):
        wait_for(self.page_has_loaded)
```

```python
html1=[]
nextFlag = True
counter = 0
while nextFlag:
    try:
        html1.append(driver.page_source)
        driver.implicitly_wait(10)
        if counter == 0:
            aa = driver.find_element_by_xpath("//table[@id='SearchResults1_dgResults']/tbody/tr[1]/td[1]/a[1]")
        else:
            aa = driver.find_element_by_xpath("//table[@id='SearchResults1_dgResults']/tbody/tr[1]/td[1]/a[2]")
        print(counter)
        counter = counter + 1

        with wait_for_page_load(driver):
            driver.find_element_by_link_text('NEXT >').click()
            #driver.implicitly_wait(10)
    except:
        nextFlag = False
```

```python
counter
```

```python
#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
counter = 1
myflag = True

for myhtml in html1:
    soup = BeautifulSoup(myhtml)
    driver.implicitly_wait(10)
    searchtable =  soup.find('table', {"id" : "SearchResults1_dgResults"})
    if searchtable is not None :
        for row in searchtable.findAll("tr"):
            cells = row.findAll('td')
            #print cells[1].text


            if (len(cells)==6 and counter > 2 and cells[0].text != "#") : #Only extract table body not heading
                A.append(cells[0].text.replace("\n", ""))
                B.append(cells[1].text.replace("\n", "").replace("\t", ""))
                C.append(cells[2].text.replace("\n", "").replace("\t", ""))
                D.append(cells[3].text.replace("\n", "").replace("\t", ""))
                E.append(cells[4].text.replace("\n", "").replace("\t", ""))
                F.append(cells[5].text.replace("\n", "").replace("\t", ""))
                links = cells[1].find_all('a')
                if len(links) > 0 :
                    G.append(links[0].get("href"))
                else:
                    G.append("None")
            counter = counter + 1
```

```python
counter
```

```python
import pandas as pd

df=pd.DataFrame(A,columns=['Number'])
df['Property_Address']=B
df['City']=C
df['Owner_Name']=D
df['Total_Value']=E
df['Type']=F
df['URL']=G

df
```
